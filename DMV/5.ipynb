```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"AirQuality.csv", sep=';')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['T'], label='NOx(GT)', color='blue')
plt.title('Air Quality Index (AQI) Over Time')
plt.xlabel('Date')
plt.ylabel('T')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['NOx(GT)'], label='PM2.5', color='red')
plt.plot(df['Date'], df['PT08.S1(CO)'], label='PM10', color='green')
plt.title('PM2.5 and PM10 Levels Over Time')
plt.xlabel('Date')
plt.ylabel('Concentration (µg/m³)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
